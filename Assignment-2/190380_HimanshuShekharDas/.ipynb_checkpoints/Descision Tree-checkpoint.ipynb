```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
%config InlineBackend.figure_format = 'retina'
```

```python
train=pd.read_csv('../data/train.csv')
train=pd.get_dummies(train,columns=["Sex","Pclass"],dtype=int)
train['Age'] = train['Age'].fillna(value = train['Age'].median())
train.describe()
```

```python
cols=train.corr().nlargest(len(train),'Survived')['Survived'].index
cm=train[cols].corr()
plt.figure(figsize=(10,5))
sns.heatmap(cm,annot=True,cmap='viridis')
```

```python
y=np.array(train["Survived"])
x=np.array(train.drop(["Survived","PassengerId"],axis=1))
```

```python
def entropy(a):
    m=len(a)
    n=np.sum(a)
    if n==0 or n==m:
        return 0,n
    else:
        p1=n/m
        p2=n/m
        return -(p1*np.log2(p1)+p2*np.log2(p2)),n
def split_entropy(y_pred,y_real):
    m=len(y_real)
    st,nt=entropy(y_real[y_pred])
    sf,nf=entropy(y_real[~y_pred])
    return (st*nt+sf*nf)/m
    
```

```python
class DTC(object):
    def __init__(self, max_depth):
        self.depth = 0
        self.max_depth = max_depth
    def best_split(self,col,y):
        min_entropy=10
        m=len(y)
        for c in col:
            y_pred=col<c
            curr_entropy=split_entropy(y_pred,y)
            if curr_entropy<=min_entropy:
                min_entropy=curr_entropy
                cutoff=c
        return min_entropy,cutoff
    def split(self,x,y):
        col=None
        min_e=1
        t=0
        for i,c in enumerate(x.T):
            e,curr_t=self.best_split(c,y)
            if e==0:
                return i,t,e
            elif e<min_e:
                min_e=e
                col=i
                t=curr_t
        return col,t,min_e

    def fit(self,x,y,node,depth=0):
        if node is None:
            return None
        if len(y)==0:
            return None
        if np.sum(y)==len(y):
            return {'val':y[0]}
        if depth>=self.max_depth:
            return None
        col,threshold,entropy=self.split(x,y)
        y_left=y[x[:,col]<threshold]
        x_left=x[x[:,col]<threshold]
        y_right=y[x[:,col]>=threshold]
        x_right=x[x[:,col]>=threshold]
        node={'col_index':col,'threshold':threshold,'val':np.round(np.mean(y))}
        node['left']=self.fit(x_left,y_left,{},depth+1)
        node['right']=self.fit(x_right,y_right,{},depth+1)
        self.depth+=1
        self.trees=node
        return node
```

```python
def prediction(node,row):
    cur_layer=node
    while cur_layer.get('left'):
        if row[cur_layer['col_index']]<cur_layer['threshold']:
            cur_layer=cur_layer['left']
        else:
            cur_layer=cur_layer['right']
    return cur_layer.get('val')

def predict(node,x):
    result=np.zeros(len(x))
    for i,c in enumerate(x):
        result[i]=prediction(node,c)
    return result
```

```python
test=pd.read_csv('../data/test.csv')
test=pd.get_dummies(test,columns=["Sex","Pclass"],dtype=int)
test['Age'] = test['Age'].fillna(value = test['Age'].median())
test_y=np.array(test["Survived"])
test_x=np.array(test.drop(["Survived","PassengerId"],axis=1))
```

```python
def f1_score(y_predict,true_y):
    tp=np.sum(y_predict * true_y)
    precision=tp/np.sum(y_predict)
    recall=tp/np.sum(true_y)
    f1=2*tp/(np.sum(true_y)+np.sum(y_predict))
    print(f"Precision= {precision:.3f}")
    print(f"Recall= {recall:.3f}")
    print(f"F1 Score= {f1:.3f}")
```

```python
def accuracy(y_predict,true_y):
    num_correct_pred=np.sum(y_predict==true_y)
    return num_correct_pred,num_correct_pred*100/y_predict.shape[0]
```

```python
dtc=DTC(max_depth=10)
m=dtc.fit(x,y,{})
test_result=predict(dtc.trees,test_x)
correct_pred=np.sum(test_result==test_y)
accuracy=correct_pred*100/test_result.shape[0]
print(f"Test Accuracy = {accuracy:.2f}%")
f1_score(test_result,test_y)
print(f"Loss={test_y.shape[0]-correct_pred}")
```

```python
train_result=predict(dtc.trees,x)
correct_pred=np.sum(train_result==y)
accuracy=correct_pred*100/train_result.shape[0]
print(f"Train Accuracy = {accuracy:.2f}%")
f1_score(train_result,y)
print(f"Loss={y.shape[0]-correct_pred}")
```
