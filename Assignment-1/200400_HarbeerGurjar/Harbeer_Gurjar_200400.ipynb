```python
##IMPLEMENTING LOGISTIC REGRESSION
##source used for help:-- https://towardsdatascience.com/logistic-regression-from-scratch-in-python-ec66603592e2
X=training_data.drop('Survived',axis=1)
y=training_data['Survived']
lr=0.0001
epochs=50000
X1=testing_data.drop('Survived',axis=1)
y1=testing_data['Survived']
```

```python
def sigmoid(z):
  return 1.0/(1+np.exp(-z))
```

```python
def loss(y, yprob):
    return -np.mean(y*(np.log(yprob)) - (1-y)*np.log(1-yprob))
   
```

```python
def gradients(X, y, yprob):
  m=X.shape[0]
  dw=(1/m)*np.dot(X.T,(yprob-y))
  db=(1/m)*np.sum((yprob-y))
  return dw,db
```

```python
def train(X,y,epochs,lr):
  m,n=X.shape
  w=np.zeros(n)
  b=0
  
  
 
  for epoch in range(epochs):
      yprob=sigmoid(np.dot(X,w)+b)
      dw,db=gradients(X,y,yprob)
      w-=lr*dw
      b-=lr*db
    
      if epoch%15==0:
        print("Loss at ",epoch," :" ,loss(y,yprob))
    
  return w,b
```

```python
def predict(X1,w,b):
  
  preds = sigmoid(np.dot(X1, w) + b)
  pred_class = []
  pred_class = [1 if i > 0.5 else 0 for i in preds]
    
  return np.array(pred_class)
```

```python
w,b=(train(X,y,epochs,lr))
```

```python
y2=predict(X1,w,b)
```

```python
 accuracy = np.sum(y1 == y2) / len(y1)
  
print(accuracy)
```
